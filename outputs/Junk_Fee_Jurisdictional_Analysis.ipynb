## 4. Chicago Neighborhood Inequality Analysis

**Data limitation:** Chicago does not collect neighborhood-level junk fee data. The Rental Registry (PRO Component 02) should be designed to fill this gap. The estimates below apply the Urban Institute's documented 10-30% fee-burden range to ACS 2022 median rents, adjusted for estimated corporate landlord concentration by community area.

```python
# Chicago community area data (ACS 2022 median rent, approximate renter HH counts)
chicago = pd.DataFrame([
    {'area':'Rogers Park','rent':1050,'renters':22800,'corp':'Med'},
    {'area':'Uptown','rent':1180,'renters':18500,'corp':'High'},
    {'area':'Lake View','rent':1580,'renters':38200,'corp':'High'},
    {'area':'Lincoln Park','rent':1750,'renters':28300,'corp':'High'},
    {'area':'Near North Side','rent':2100,'renters':32000,'corp':'High'},
    {'area':'Loop','rent':2200,'renters':15200,'corp':'High'},
    {'area':'Near South Side','rent':1950,'renters':12800,'corp':'High'},
    {'area':'Near West Side','rent':1650,'renters':18500,'corp':'High'},
    {'area':'West Town','rent':1550,'renters':24000,'corp':'High'},
    {'area':'Logan Square','rent':1350,'renters':22000,'corp':'High'},
    {'area':'Edgewater','rent':1150,'renters':17500,'corp':'High'},
    {'area':'Hyde Park','rent':1300,'renters':10200,'corp':'High'},
    {'area':'Douglas','rent':1100,'renters':8200,'corp':'High'},
    {'area':'Humboldt Park','rent':950,'renters':15800,'corp':'Med'},
    {'area':'Austin','rent':850,'renters':19200,'corp':'Med'},
    {'area':'South Shore','rent':870,'renters':14200,'corp':'Med'},
    {'area':'Englewood','rent':780,'renters':6800,'corp':'Low'},
    {'area':'West Englewood','rent':800,'renters':6400,'corp':'Low'},
    {'area':'North Lawndale','rent':830,'renters':7200,'corp':'Low'},
    {'area':'West Garfield Park','rent':780,'renters':4800,'corp':'Low'},
    {'area':'East Garfield Park','rent':820,'renters':5200,'corp':'Med'},
    {'area':'Chatham','rent':900,'renters':10800,'corp':'Low'},
    {'area':'Auburn Gresham','rent':880,'renters':8200,'corp':'Low'},
    {'area':'Chicago Lawn','rent':900,'renters':10400,'corp':'Low'},
    {'area':'Roseland','rent':900,'renters':8200,'corp':'Low'},
    {'area':'Woodlawn','rent':950,'renters':8600,'corp':'Med'},
    {'area':'Greater Grand Crossing','rent':850,'renters':7800,'corp':'Low'},
    {'area':'Bridgeport','rent':1100,'renters':8200,'corp':'Med'},
    {'area':'Albany Park','rent':1000,'renters':14500,'corp':'Med'},
    {'area':'Brighton Park','rent':900,'renters':8800,'corp':'Low'},
    {'area':'Belmont Cragin','rent':1000,'renters':14800,'corp':'Low'},
    {'area':'South Lawndale','rent':850,'renters':12800,'corp':'Low'},
    {'area':'Avondale','rent':1200,'renters':11200,'corp':'Med'},
    {'area':'Riverdale','rent':700,'renters':2200,'corp':'Low'},
    {'area':'Beverly','rent':1350,'renters':4200,'corp':'Low'},
])

# Fee burden calculation
def fee_rate(corp):
    return {'High': 0.25, 'Med': 0.18, 'Low': 0.12}[corp]

chicago['fee_rate'] = chicago['corp'].apply(fee_rate)
chicago['est_monthly_fee'] = (chicago['rent'] * chicago['fee_rate']).round(0).astype(int)
chicago['est_annual_fee'] = chicago['est_monthly_fee'] * 12
chicago['est_income'] = (chicago['rent'] * 12 / 0.30).round(0).astype(int)  # 30% rent-to-income
chicago['fee_pct_income'] = ((chicago['est_annual_fee'] / chicago['est_income']) * 100).round(1)
chicago['total_hh_burden'] = (chicago['est_annual_fee'] * chicago['renters']).round(0).astype(int)

print('CHICAGO COMMUNITY AREAS: ESTIMATED JUNK FEE BURDEN')
print('(35 selected areas shown; 77 total in full dataset)')
print()
display(chicago[['area','rent','renters','corp','est_monthly_fee','est_annual_fee','fee_pct_income']]
        .sort_values('est_annual_fee', ascending=False)
        .rename(columns={
            'area':'Community Area', 'rent':'Median Rent',
            'renters':'Renter HH', 'corp':'Corp. Landlord',
            'est_monthly_fee':'Est. Monthly Fee',
            'est_annual_fee':'Est. Annual Fee',
            'fee_pct_income':'Fee as % Income'
        }))
```

## 6. Recommended Policy Design (Option E: Layered)

1. **Enumerated Fee Allowlist** (Oregon + Olympia model): Only listed fees permitted; everything else banned by default.

4. **Rent-Increase Frequency Limit**: Max one increase per 12-month cycle, 90-day notice. Manages pass-through risk.
